```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten, Conv2D, MaxPooling2D
```

```python
models = Sequential()

# 1st conv2D Layer
models.add(Conv2D(96, kernel_size = (11, 11), strides= (4,4), padding="valid",activation= 'relu', input_shape = (227,227, 3)))

models.add(MaxPooling2D(pool_size=(3,3), strides = (2, 2), padding = "valid", data_format= None))           






# 2nd  conv2D Layer
models.add(Conv2D(256, kernel_size = (5, 5), strides= 1, padding="same",activation= 'relu'))

models.add(MaxPooling2D(pool_size=(3,3), strides = (2, 2), padding = "valid", data_format= None))           





# 3rd con2D Layer
models.add(Conv2D(384, kernel_size = (5, 5), strides= 1, padding="same",activation= 'relu'))



# 4th con2D Layer
models.add(Conv2D(384, kernel_size = (5, 5), strides= 1, padding="same",activation= 'relu'))




# 5th con2D Layer
models.add(Conv2D(384, kernel_size = (5, 5), strides= 1, padding="same",activation= 'relu'))

           
# Flatten layer
models.add(Flatten())
           
models.add(Dense(4096, activation ='relu'))
models.add(Dense(4096, activation ='relu'))           
           
models.add(Dense(2, activation ='softmax'))
           
models.compile(loss = "categorical_crossentorpy", optimizer = 'adam', metrics=['accuracy'])
```
